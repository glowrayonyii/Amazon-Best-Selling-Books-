### DATA CLEANING 

This process is very crucial in Data curation. So to get a quality data, we will be cleaning our dataset by filling and dropping missing values, columns or rows, stripping off white spaces and changing data types. 

```python
#drop the Unnamed and Ranks column
df_books.drop(columns = ['Unnamed: 0', 'Ranks'], inplace = True)
df_books.head()
```

```python
#change the price, year and ratings column data type to str so as to strip whitespaces
df_books['Price'] = df_books['Price'].astype(str)
df_books['Ratings'] = df_books['Ratings'].astype(str)
df_books['Year'] = df_books['Year'].astype(str)
df_books.info()
```

```python
#strip off whitespaces
df_books['Title'].str.strip()
df_books['Authors'].str.strip()
df_books['Book_cover'].str.strip()
df_books['Price'].str.strip()
df_books['Reviews'].str.strip()
df_books['Ratings'].str.strip()
df_books['Year'].str.strip()
df_books
```

```python
# Identify rows where Review is not NaN and Authors is NaN
mask_off = (df_books['Reviews'] != 'NaN') & (df_books['Authors'].isna())

# Swap the values between Reviews and Authors
df_books.loc[mask_off, ['Reviews', 'Authors']] = df_books.loc[mask_off, ['Authors', 'Reviews']].values
df_books.head(500)
```

### Q6: Top 5 authors each year?

```python
top_authors = df_books.groupby('Year')['Authors'].value_counts().reset_index(name = 'Count')
years = top_authors['Year'].unique()
fig, axes = plt.subplots(nrows=len(years), ncols=1, figsize=(8, 4*len(years)))

for i, year in enumerate(years):
    top_authors_year = top_authors[top_authors['Year'] == year].head(5)
    axes[i].barh(top_authors_year['Authors'], top_authors_year['Count'])
    axes[i].set_title(f"Top 5 Authors in {year}")
    axes[i].set_xlabel('Authors')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

### Q7: what were the top 10 most reviewed books?

```python
top_review = df_books.groupby('Title')['Reviews'].sum().nlargest(10).sort_values(ascending = True).reset_index(name='Reviews')
top_review
plt.figure(figsize=[8, 6])
plt.barh(top_review['Title'], top_review['Reviews'], color = 'blue')
plt.xlabel('Reviews')
plt.ylabel('Title')
plt.title('Top 10 mostly reviewed books')
plt.savefig('bookreview.png', bbox_inches='tight', dpi=300)
plt.show()
print(top_review)
```
